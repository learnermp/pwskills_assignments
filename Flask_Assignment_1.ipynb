Q1. What is Flask Framework? What are the advantages of Flask Framework?

Ans:

Flask is python based web framework for making web apps.

Advantages:

1) Flask is lightweight and easy to comprehend web framework.

2) It supports intergration of HTML and CSS.

3) ML/Deep learning models can be deployed using Flask.

4) It is open sourced and has large developer community.

5) It provides easy testing and debugging.

Q3. What is App routing in Flask? Why do we use app routes?

Ans:

App routing is a decorator that mapps URL to a function in the FLASK app. In other words, it routs the function to the specific route. In the previous example, @app.route('/') maps hello_world() to home url (127.0.0.0:5000)

Key reasons to use @app.route are ⁉

1) URL mapping

2) for taking http requests methods of web api (POST, GET, PUT, DROP)

3) It is easy to use and implement
