example : expanding

dataframe['column'].expanding(min_periods = 2).sum() first calculates the sum of first two values. Next time it slides down one row and finds sum for two values and the previous sum. This gradually expands each time it slides down the row by incorporating previous sum for calculating new sum.

Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and
calculates the difference between them in days, hours, and minutes using Pandas time delta. The
program should prompt the user to enter the dates and display the result.

Ans:

```python
def date_diff():

  date1 = input("Enter the first date in 'YYYY-MM-DD' format: ")
  date2 = input("Enter the second date (later) in 'YYYY-MM-DD' format: ")

  dt1 = pd.to_datetime(date1)
  dt2 = pd.to_datetime(date2)

  difference = pd.Timedelta(dt2 - dt1)

  days = difference.days
  hours = difference.components.hours
  minutes = difference.components.minutes

  return days, hours, minutes
```

Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
column to a categorical data type. The program should prompt the user to enter the file path, column
name, and category order, and then display the sorted data.

Ans:

```python
import pandas as pd
def col_to_categorical():

    file_path = input("Provide the file path for csv file: ")
    col_name = input("Enter the column name to convert into categorical data type: ")
    cat_order = input("Enter the category order separted by commas without any spaces in-between: ").split(",")


    data = pd.read_csv(file_path)
    data[col_name] = pd.Categorical(data[col_name], categories=cat_order, ordered=True)

    return data.sort_values(by=[col_name])
```

Q10. Write a Python program that reads a CSV file containing sales data for different products and
visualizes the data using a stacked bar chart to show the sales of each product category over time. The
program should prompt the user to enter the file path and display the chart.

Ans:

```python
import pandas as pd

def sales_bar():

  file_path = input("Enter the file path: ")

  df = pd.read_csv(file_path)

  pivot_table = df.pivot_table(index='product', columns='moth', values='sale', aggfunc='sum')

  pivot_table.plot(kind='bar', stacked=True, figsize=(10, 7))
```
