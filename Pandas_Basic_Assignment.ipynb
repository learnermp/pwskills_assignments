Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

Ans:

Pandas Series and DataFrame are mutable while Panel is immutable 3D data format.

Example of mutability of DataFrame:

```python
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Modify the DataFrame
df['A'] = [7, 8, 9]
df.loc[2, 'B'] = 1000

print("\nModified DataFrame:")
print(df)
```

Example for mutability of Series:

```python
data = [1, 2, 3]
series = pd.Series(data)

print("Original Series:")
print(series)

# Modify the Series, Series is like a list follows indexing like former
series[0] = 4
series[2] = 6

print("\nModified Series:")
print(series)
```
